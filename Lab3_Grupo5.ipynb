```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/Olympics2020.csv')
data=pd.DataFrame(df)
print(data)
#print(data.columns)
#dataFiltered=data.loc[(data[' Name']==" Viktor Minibayev")]
#dataFiltered
```

```python
#Filtro por país: Estados Unidos y deporte: Buceo
dataFilteredUSA=data.loc[(data[' Team']==" United States") & (data[' Sport']==" Diving")]
dataFilteredUSA
```

```python
#Filtro por país: China y deporte: buceo
dataFilteredCHN=data.loc[(data[' Team']==" China") & (data[' Sport']==" Diving")]
dataFilteredCHN
```

```python
#Edades de menor a mayor entre las personas participantes en el deporte de buceo en Estados Unidos
print(dataFilteredUSA.sort_values(by=[' Age'], ascending=True))
```
